```python
import numpy as np
import re

#Function to clean the Time column
def clean_time(value):
    if pd.isna(value):
        return np.nan
    value = str(value).strip().lower().replace('"', '').replace("'", "")

    #Turning key words to specific times
    if 'morning' in value:
        return '09:00'
    if 'afternoon' in value:
        return '15:00'
    if 'evening' in value or 'dusk' in value or 'sunset' in value:
        return '19:00'
    if 'night' in value:
        return '23:00'
    if 'noon' in value or 'midday' in value or 'lunchtime' in value:
        return '12:00'
    if 'dawn' in value or 'daybreak' in value:
        return '06:00'
    if 'midnight' in value:
        return '00:00'
    if 'unknown' in value or 'not stated' in value or 'not advised' in value or value in ['?', '', 'nan', '--', '\xa0', 'x', 'n']:
        return np.nan

    #Looking for patterns like 14h30, 14h, 14:30, 1430, 14hoo, 2pm, 2 am, etc and turn these into HH:MM format
    match = re.search(r'(\d{1,2})h(\d{2})', value)
    if match:
        return f"{int(match.group(1)):02d}:{match.group(2)}"
    match = re.search(r'(\d{1,2})h', value)
    if match:
        return f"{int(match.group(1)):02d}:00"
    match = re.search(r'(\d{1,2}):(\d{2})', value)
    if match:
        return f"{int(match.group(1)):02d}:{int(match.group(2)):02d}"
    match = re.search(r'(\d{1,2})(\d{2})\s*hrs?', value)
    if match:
        return f"{int(match.group(1)):02d}:{int(match.group(2)):02d}"
    match = re.search(r'(\d{1,2})hrs(\d{2})', value)
    if match:
        return f"{int(match.group(1)):02d}:{match.group(2)}"
    match = re.search(r'(\d{1,2})(\d{2})$', value)
    if match:
        return f"{int(match.group(1)):02d}:{match.group(2)}"
    match = re.search(r'(\d{1,2})j(\d{2})', value)
    if match:
        return f"{int(match.group(1)):02d}:{match.group(2)}"
    match = re.search(r'(\d{1,2})hoo', value)
    if match:
        return f"{int(match.group(1)):02d}:00"
    match = re.search(r'(\d{1,2})\s*(am|pm)', value)
   
   #If pattern matches, convert to 24-hour format
    if match:
        hour = int(match.group(1))
        if match.group(2) == 'pm' and hour != 12:
            hour += 12
        if match.group(2) == 'am' and hour == 12:
            hour = 0
        return f"{hour:02d}:00"
    
    #If no pattern matched, return NaN
    return np.nan

#Create the new column Time_cleaned applying the function clean_time to the Time column
df['Time_cleaned'] = df['Time'].apply(clean_time)
print(df['Time_cleaned'].isna().sum())
df
```

```python
#Grouping the Time_cleaned column into time ranges
def time_group(time_str):
    if pd.isna(time_str):
        return 'NaN'
    hour = int(time_str.split(':')[0])
    if 0 <= hour < 6:
        return 'Late night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 14:
        return 'Midday'
    elif 14 <= hour < 19:
        return 'Afternoon'
    elif 19 <= hour < 24:
        return 'Night'
    else:
        return 'NaN'

#Create the new column Time_cleaned_group applying the function time_group to the Time_cleaned column
df['Time_cleaned_group'] = df['Time_cleaned'].apply(time_group)
print(df['Time_cleaned_group'].value_counts())
```

```python
#Create a bar chart to visualize the number of shark attacks in each time group
import matplotlib.pyplot as plt

#Order the categories for better visualization
order = ['Morning', 'Midday', 'Afternoon', 'Night','Late night', 'NaN']
df['Time_cleaned_group'] = pd.Categorical(df['Time_cleaned_group'], categories=order, ordered=True)
ax = df['Time_cleaned_group'].value_counts().loc[order].plot(kind='bar', color='skyblue')
plt.xlabel('Time Group')
plt.ylabel('Number of Attacks')
plt.title('Shark Attacks by Time of Day')
plt.xticks(rotation=45)

#Add value labels on top of each bar in the surfing activities bar chart
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')
```

```python
#Create a bar chart to visualize the number of shark attacks in each time group for Surfing activities
import matplotlib.pyplot as plt

#Filter only the rows where the 'Surfing' column is True
df_surfing = df[df['Surfing'] == True]

#Order the categories for better visualization
order = ['Morning', 'Midday', 'Afternoon', 'Night', 'Late night', 'NaN']
df_surfing['Time_cleaned_group'] = pd.Categorical(df_surfing['Time_cleaned_group'], categories=order, ordered=True)

ax = df_surfing['Time_cleaned_group'].value_counts().loc[order].plot(kind='bar', color='skyblue')
plt.xlabel('Time Group')
plt.ylabel('Number of Surfing Attacks')
plt.title('Shark Attacks by Time of Day (Surfing Activities)')
plt.xticks(rotation=45)

#Add value labels on top of each bar in the surfing activities bar chart
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')

plt.show()
```
